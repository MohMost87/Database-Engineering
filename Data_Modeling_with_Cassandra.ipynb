**Combine all files in one new file**

```python
import os
import glob
import csv
import pandas as pd

# folder path that contain all csv files
filepath = '/content/sample_data/Event_data'
# check for all paths


file_path_list = sorted(glob.glob(os.path.join(filepath, '*.csv')))
print(f"{len(file_path_list)} ")

# to store the rows create a list
full_data_rows_list = []

# Read each file to extract data in them as rows
for f in file_path_list:
    with open(f, 'r', encoding = 'utf8', newline='') as csvfile:
        csvreader = csv.reader(csvfile)
        next(csvreader) # this to pass header
        for line in csvreader:
            full_data_rows_list.append(line)

print(f"إTotal number of rows {len(full_data_rows_list)}")

# create standard columns
csv.register_dialect('myDialect', quoting=csv.QUOTE_ALL, skipinitialspace=True)

output_filename = 'event_datafile_new.csv'

with open(output_filename, 'w', encoding = 'utf8', newline='') as f:
    writer = csv.writer(f, dialect='myDialect')
    # create columns that specified to project
    writer.writerow(['artist','firstName','gender','itemInSession','lastName','length',\
                'level','location','sessionId','song','userId'])

    #
    for row in full_data_rows_list:
        # check for next line if it empty
        if (row[0] == '' or row[10] != 'NextSong'):
            continue
        # we read each row in wight order
        writer.writerow([row[0], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[12], row[13], row[16]])

print(f" : {output_filename}")
```
